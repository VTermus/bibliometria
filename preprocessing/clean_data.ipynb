```python
# duplicates by title
duplicates = df[df.duplicated('title', keep=False)]
print(df.duplicated('title').sum())

duplicates_sorted = duplicates.sort_values('title')
print(duplicates_sorted[['title', 'issn', 'sjr_best_quartile', 'sjr']].to_string())
```

# Web of science

```python
df = pd.read_csv('wos_journals.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
# full duplicates
df.duplicated().any()
```

```python
# duplicates except 'wos_index'
has_duplicates = df.duplicated(subset=df.columns.difference(['wos_index']), keep=False).any()
has_duplicates
```

```python
# duplicates except 'wos_index' and 'url'
other_columns = df.columns.difference(['wos_index', 'url']).tolist()

duplicates_mask = df.duplicated(subset=other_columns, keep=False)

if duplicates_mask.any():
    print(duplicates_mask.sum())

    duplicate_groups = df[duplicates_mask].groupby(other_columns)
    for name, group in list(duplicate_groups)[:3]:  
        if len(group) > 1:
            print(group)
```

```python
# duplicates except 'wos_index', 'url' and 'category'
other_columns = df.columns.difference(['wos_index', 'url', 'category']).tolist()
duplicates_mask = df.duplicated(subset=other_columns, keep=False)

if duplicates_mask.any():
    print(duplicates_mask.sum())
```

```python
# duplicates by title
duplicates = df[df.duplicated('title', keep=False)]
print(df.duplicated('title').sum())

duplicates_sorted = duplicates.sort_values('title')
#print(duplicates_sorted[['title', 'issn', 'eissn', 'wos_index', 'jif']].to_string())
```
